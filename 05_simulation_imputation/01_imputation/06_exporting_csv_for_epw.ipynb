```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from dateutil.parser import parse
```

```python
imputed_file = '../../01_weather_data/03_imputing_process/02_imputed_data/tmx_46days_imputed'#01_test_data/tmx_base_2018_solar_angles.csv'#02_imputed_data/Tmx_CNN-LSTM_imputed.csv'#'../01_data/01_PreEPW/imputed_SARIMA_0_1_1.csv"
to_epw_file= '../../01_weather_data/04_to_EPWs/imputed_46days.csv' #base_epw_adiabatic.csv
epw = pd.read_csv(imputed_file,
                 index_col=0,
                 parse_dates=True)
epw.fillna(0,inplace=True)
```

```python
epw
```

```python
epw.plot(subplots=True,figsize=(20,40));
```

```python
def df2epw(df):
    df['anio']    = df.index.year
    df['mes']     = df.index.month
    df['dia']     = df.index.day
    df['hora']    = df.index.hour
    df['minutos'] = df.index.minute
    return df[1:]
```

```python

columnas = ['anio','mes','dia','hora','minutos',
            'to','RH','P',"Ig","Ib",'WD','WS']
tmp = df2epw(epw)

tmp.to_csv(to_epw_file,
           columns=columnas,header=False,index=False)
```
