```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ann=pd.read_csv('../../01_weather_data/03_imputing_process/02_imputed_data/Tmx_CNN-LSTM_imputed.csv',index_col=0,parse_dates=True)
sarima=pd.read_csv('../../01_weather_data/03_imputing_process/02_imputed_data/Tmx_SARIMA_mae_0_1_0_0_1_1.csv',index_col=0,parse_dates=True)
complete=pd.read_csv('../../01_weather_data/03_imputing_process/01_test_data/base_temixco.csv',index_col=0,parse_dates=True)
```

```python
ann['mae']=np.abs(ann.Ig-complete.Ig)
sarima['mae']=np.abs(sarima.Ig-complete.Ig)
ann.head()
```

```python
complete.head()
```

```python
dias=['2018-01-12',
 '2018-01-30',
 '2018-04-17',
 '2018-06-15',
 '2018-07-19',
 '2018-08-12',
 '2018-11-14',
 '2018-12-21']
```

```python
sarima
```

```python
ann_mae=ann.mae.resample('D').mean()
sarima_mae=sarima.mae.resample('D').mean()
```

```python
ann_maes=[]
sarima_maes=[]
for i in range(len(dias)):
    fig,ax=plt.subplots(figsize=(20,5))
    print('ann_mae:',ann_mae[dias[i]])
    ann_maes.append(ann_mae[dias[i]])
    print('sarima_mae:',sarima_mae[dias[i]])
    sarima_maes.append(sarima_mae[dias[i]])                       
    ax.plot(ann.Ig.loc[dias[i]],'g.-',label='CNN-LSTM_imputed')
    ax.plot(sarima.Ig.loc[dias[i]],'r.-',label='SARIMA_imputed')
    ax.plot(complete.Ig.loc[dias[i]],'k-',label='Measured')
    ax.legend()
    ax.set_ylabel('$I_g[\dfrac{W}{m^2}]$')
    ax.set_xlabel('Date[mm-dd-HH]')
    plt.show()
```

```python
dates=pd.DataFrame(dias)
sarima_error=pd.DataFrame(sarima_maes)
ann_error=pd.DataFrame(ann_maes)
ann_error
```

```python
errors=pd.concat((dates,sarima_error,ann_error),axis=1,keys=['dates','sarima_mae','ann_mae'])
```
